```python
import skimage.io as io
import plotly.graph_objects as go
import numpy as np
```

```python
sync = io.imread('/Users/jefft/Desktop/Chan lab/SRTP/ImageAnalysis/PCNAdeep/ccnbDeep/example/out/synchrogram/trk-mt_track_11.tif')
sync = sync/65535
```

```python
from skimage.util import img_as_ubyte
sync = img_as_ubyte(sync)
```

```python
frame = 17
fig = go.Figure(data=[go.Surface(z=sync[frame,:,:,1], colorscale='Earth')])

fig.update_layout(title='GFP frame '+str(frame), autosize=True)

fig.show()
```
